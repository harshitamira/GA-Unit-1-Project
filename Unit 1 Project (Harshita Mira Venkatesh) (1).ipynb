```python

# Challenge 1: A palindromic number reads the same both ways. For example, 1234321 is a palindrome. 
#             The largest palindrome made from the product of two two-digit numbers is 9009 = 91 × 99. 
#             Find the largest palindrome made from the product of two three-digit numbers. 
#             Afterward, write a brief explanation walking through your code's logic in markdown.
 
# Initiate all lists and import packages
import numpy as np
import pandas as pd
three_digit_int = []
new_three_dig_int = []
possible_palindromes = []
new_possible_palindromes = []


# Step 1: Construct a list of all three digit numbers and multiply each 
#         element by every other element to get the universe of all three digit numbers possible:

for x in range(100, 1000): # the range of the smallest and largest number I can get by mutliplying two 3-digit numbers
    for i in range(100, 1000):
        product = x*i # universe of every possible 3-digit * 3-digit product
        three_digit_int.append(product) #

for x in three_digit_int: # convert to string in order to subscript and avoide an error.
    x = str(x)
    new_three_dig_int.append(x)


# Step 2: Here I insert the logic of the palindrome. I know for a fact that that the smallest
#         number I can get by multiplying 3 digits times three digits (100 x 100) = 10,000 - a five digit number.
#         and the largest number I can get by multiplying 3 digits times three digits (999 x 999) = 998,001 - a six digit number
#         so the required palindrome is between five and six digits. Using this logic and my understanding of the 
#         general structure of a palindrome, I construct the code:
    
for x in new_three_dig_int:
    if len(x)==5:
        if x[0]==x[4] and x[1]==x[3]: # the first digit equals the last digit and the 10th digit equals the thousandth digit, in a 4-digit palindrome
            possible_palindromes.append(x)
    elif len(x)==6:
        if x[0]==x[5] and x[1]==x[4] and x[2] == x[3]: # the first digit equals the last digit and the 10th digit equals the ten-thousandth digit and the hundreth digit equals the thousandth digit, in a 5-digit palindrome.
            possible_palindromes.append(x)
            
for x in possible_palindromes:
    x = int(x) # convert back into integer to take the max in the next step
    new_possible_palindromes.append(x)


# Step 3: Print out the maximum of the universe of palindromes constructed by multiplying 3-digits by 3-digits.
print("Max 3-digit times 3-digit palindrome:", max(new_possible_palindromes)) 
    


```

```python
# Challenge # 2: The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17. 
# Find the sum of all the primes below 2,000. Afterward, write a brief explanation walking 
# through your code's logic in markdown.


#initiate empty lists
prime_list = []

# Step 1: Create a function to test whether a number is prime or not. A number is considered
#         to be a prime number if it can be divided by ONLY 1 and itself, I use this logic
#         to construct the function below:

def test_prime(number):
    if number == 1: # 1 is neither prime nor composite, so this function should return False if 1 is entered.
        return False
    
    for x in range(2, number): # I test divisibility by all numbers from 2 to (nunmber - 1), if it can't, 
                                # be divided by any of these number then it is prime.
        if number % x == 0: 
            return False
            
# Step 2: # I loop through all numbers from 1-1999 (all numbers less than 2000) and test if they are prime
         
for i in range(1, 2000): 
    if test_prime(i) != False: # test if a number is prime by calling the above function.
        prime_list.append(i)   # if a number is prime I append it to an empty list.

# Step 3: Print result by summing all elements of the list prime_list:
print("Sum of prime numbers below 2000 is:", sum(prime_list))
            
            
```

```python
# Challenge 3: If we list all of the natural numbers below 10 that are multiples of 3 or 5, 
            # we get 3, 5, 6, and 9. The sum of these multiples is 23. Find the sum of all the multiples 
            # of 3 and 5 below 1,000. Afterward, write a brief explanation walking through your code's logic in markdown.

# Initialize lists and sets
three_five_multiples = []

# Step 1: Create a function testing whether am input is a multiple of 3 or 5

def is_threefive_multiple(num):
    if num % 5 == 0 or num % 3 == 0: # if a number is a multiple of a second number, it returns a 0 remainder when divided by the second number
        return True

# Step 2: Loop through all the numbers below 1000 (1-999) and test if they are multiples of 3 or 5.
#         if these numbers are multiples of 3 or 5 append them to an empty list 
        
for x in range(1,1000): 
    if is_threefive_multiple(x)==True: # testing if a multiple of 3 or 5 by calling the function.
        three_five_multiples.append(x) # appending 3/5 multiples to an empty list

# Step 4: Sum all the itemps of the list created in the previpus step and print the answer.
print("The sum of multiples of 3 or 5 below 1,000 is:", sum(three_five_multiples))    
```

```python
# Challenge 4: String Compressor
            # Implement a method to perform basic string compression using the counts of repeated characters. 
            # (This is called run-length encoding.) For example, the string "aabcccccaaa" would become a2b1c5a3. 
            # If the “compressed” string would not become smaller than the original string, your method should 
            # return the original string. You can assume the string has only uppercase and lowercase letters (a–z). 
            # Specify whether your solution is case sensitive or case insensitive and what you would need to change 
            # to make it the other. Afterward, write a brief explanation walking through your code's logic in markdown.


# Design compress function
            
def compress_fn(string):
    compress_string = "" # initialize compress string. Set to be empty for now.
    
    start_letter = "" # when we start couting, the start letter will be every point where you start counting a substring 
                        # of same letters. 
                        # for e.g., in "aaabbbcc", the first a, first b and first c are the start_letter
    counter = 1         # create a counter to keep count of the number of iterations the same letter appears before the next letter.
                        # given how the code is constructed, we start the counter at 1, so that the first letter in a substring is counted.
    
    for letter in string:
        
        # first part of if-else statement tests if a letter is not equal to the prebious letter.
        # so for e.g, in "AABBB", it is at the poinf where the substring of As end and Bs begin
        
        if letter != start_letter:
            if start_letter: # short hand of "if start_letter not equal to empty string"
                compress_string += "{}{}".format(start_letter, counter) # format substring compression.
            start_letter = letter # set start letter to a the letter we are on. F
            counter = 1 # since the letter and the previous letter aren't the same letter the counter sets back to 1.
       
       # second part of the if-else statement: # if a letter and the previous letter are the same, 
       # the counter gets incremented,# for e.g, in "aaa", the counter goes from 1, to 2, to 3.

        else:             
            counter += 1
            
    compress_string += "{}{}".format(start_letter, counter) # format substring compression.
    
    # Test whether the compressed string is actually shorter.
    if len(compress_string)>=len(string): # if the compressed string isn't shorter, return the original string.
        return(string)
    else:
        return(compress_string) # otherwise return the compressed string.


#Test CASE-SENSITIVE VERSION of compress_string here           
print("Tested case-sensitive compress function output:", compress_fn("abBBggggggc"))
```
